## Understanding the Basics of Image

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import skimage
from skimage import io,img_as_float
```

```python
#Reading a sample image file using skimage's io module
img=io.imread("sample.jpg")
img.shape,img.min(),img.max(),img.dtype
```

```python
blank=np.zeros((183,275,3),dtype='uint8')
```

```python
blank.shape
```

```python
plt.imshow(blank)
```

```python
# blank[10:200,10:200,0]=255
# blank[10:200,10:200,1]=0
# blank[10:200,10:200,2]=255


#Put box in center
blank=np.zeros((600,500,3),dtype='uint')
blank[int(blank.shape[0]/2-50):int(blank.shape[0]/2+50),200:300,:]=[255,255,0]
plt.imshow(blank)
plt.show()
```

```python
# Creating a numpy array of 500* 500 with random numbers and converting it into a image
plt.imshow(np.random.random((183,275,3)))
plt.show()
```

```python
img_float=img_as_float(img)
img_float.min(),img_float.max()
```

```python
plt.imshow(img_float)
plt.show()
```

```python
img_float.shape
```

```python
dark=(img-np.mean(img))/np.std(img)
```

```python
dark.min(),dark.max(),dark.shape
```

```python
plt.imshow(dark)
plt.show()
```
